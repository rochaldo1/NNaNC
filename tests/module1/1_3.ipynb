Задание: **Вариант 5. Инициализация матриц для SVD**

Для матрицы оценок R размером 5×5 (задайте случайно) инициализируйте матрицы P (размер 5×d) и Q (размер d×5) со случайными значениями. d = 3. 

Рассчитайте предсказанную матрицу R_pred и среднеквадратическую ошибку (MSE) между исходной R и R_pred для ненулевых элементов.

---

```python
import numpy as np

np.random.seed(42)

n_users = 5
n_items = 5
d = 3

R = np.random.randint(0, 6, size=(n_users, n_items))

print('Матрица R:')
print(R)

P = np.random.rand(n_users, d)
Q = np.random.rand(d, n_items)

print('\nМатрица P:')
print(P)
print('\nМатрица Q:')
print(Q)

R_pred = P @ Q

print('\nПредсказанная матрица R_pred:')
print(R_pred)

mask = R != 0
mse = np.mean((R[mask] - R_pred[mask]) ** 2)

print('\nMSE (по ненулевым элементам):', mse)
```
